# 对地区表进行处理

此表我们需要 provinceName，province_confirmedCount，province_curedCount，province_deadCount，updateTime 以省份作为指标

```python
# 对 provinceName 字段进行去重，选择需要的列
area_df = area_df.drop_duplicates(subset=['provinceName','updateTime'])
area_df = area_df[['provinceName','province_confirmedCount','province_curedCount',
                    'province_deadCount','updateTime']].set_index('provinceName').reset_index()
area_df.head()
```

```python
# 得到目标表后，查看有无缺失值
area_df.info()
```

## 绘制最新一天的确诊人数疫情热力地图

```python
# 选择最新一天的，也就是 2 月 15 的行，剔除 provinceName 字段中的‘省’‘市’,处理特殊区域
map_df = area_df[area_df['updateTime']=='2/15']
map_df['provinceName'] = map_df['provinceName'].str.\
    findall('(.*?)[省|市|自治区|回族自治区|维吾尔自治区|壮族自治区]').str.get(0)
map_df.head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map


provinces = list(map_df['provinceName'])
values = list(map_df['province_confirmedCount'])


def map_visualmap() -> Map:
    c = (
        Map()
        .add("", [list(z) for z in zip(provinces, values)], "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="确诊数(截至时间为2-15)"),
            visualmap_opts=opts.VisualMapOpts(max_=1500, is_piecewise=True,
                                             pieces=[
                                                 {"min": 0, "max": 300},
                                                 {"min": 300, "max": 600},
                                                 {"min": 600, "max": 900},
                                                 {"min": 900, "max": 1200},
                                                 {"min": 1200, "max": 1500},
                                                 {"min": 1500},
                                             ]),
        )
    )
    return c


c = map_visualmap()
c.render_notebook()
c.render('确诊数热力地图.html')
```

## 绘制疫情走势折线图

```python
# 以日期进行分组求和
line_df = area_df.groupby('updateTime').aggregate('sum').reset_index()
line_df = line_df.sort_values(by=['province_confirmedCount'])
line_df.head()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line


date = list(line_df.updateTime)
confirmed = list(line_df.province_confirmedCount)
cured = list(line_df.province_curedCount)
dead = list(line_df.province_deadCount)


def line_base() -> Line:
    c = (
        Line()
        .add_xaxis(date)
        .add_yaxis(
            "确诊", confirmed,color='grey'
        )
        .add_yaxis(
            "治愈", cured,color='green'
        )
        .add_yaxis(
            "死亡", dead,color='red'      
        )
        .set_global_opts(
                title_opts=opts.TitleOpts(title="疫情走势"),
        )
    )
    return c


c = line_base()
c.render_notebook()
c.render('疫情走势.html')
```

## 绘制疫情情况圆环图

```python
# 确诊 ：死亡 : 治愈
pie_df = map_df.groupby('updateTime').agg('sum')


from pyecharts import options as opts
from pyecharts.charts import Pie


counts = ['确诊','治愈','死亡']
values = [pie_df.province_confirmedCount[0],pie_df.province_curedCount[0],
         pie_df.province_deadCount[0]]


def pie_radius() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(counts, [66256, 8098, 1519])],
            radius=["55%", "70%"],
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title=""),
            legend_opts=opts.LegendOpts(
                orient="vertical", pos_top="75%", pos_left="18%"
            ),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
        .set_colors(["red", "green", "grey"])
    )
    return c


c = pie_radius()
c.render_notebook()
c.render('比例圆环图.html')
```

## 绘制疫情新闻数走势折线图

```python
# 只需要 pubDate 字段，对出现的日期次数进行计算
new_line_df = new_df
new_line_df['pubDate'] = pd.to_datetime(new_df['pubDate']).dt.normalize()
new_line_df = new_line_df.iloc[:,[2]]
new_line_df = new_line_df['pubDate'].value_counts().reset_index()
new_line_df.columns = ['date','count']
new_line_df = new_line_df.sort_values(by='date')
new_line_df = new_line_df.drop(new_line_df.index[[0]],axis=0)
new_line_df.head()
```

```python
# 绘制折线图
import pyecharts.options as opts
from pyecharts.charts import Line


time_list=[x.strftime('%m-%d') for x in new_line_df['date']]
line = Line()
line.add_xaxis(time_list)
line.add_yaxis("新闻数", list(new_line_df['count']))
line.set_global_opts(title_opts=opts.TitleOpts(title=""))
line.render_notebook()
line.render('新闻数走势.html')
```

# 对谣言表进行处理

```python
# 只需要 mainSummary，title 字段
rumous_df = rumous_df[['mainSummary','title']]
rumous_df.head()
```

## 来源饼状图

```python
# 通过分词查看来源关键词
l = ''.join(list(rumous.mainSummary))
import jieba
ls = jieba.lcut(l)
c = {}
for i in ls:
    c[i] = c.get(i,0) + 1
```

```python
# 统计来源出现次数
count = {
    '丁香医生' : 0,
    '世界卫生组织' : 0,
    '医院' : 0,
    '院士' : 0,
    '公安' : 0
}
for ru in list(rumous.mainSummary):
    if '丁香医生' in ru:
        count['丁香医生'] += 1
    if '世界卫生组织' in ru:
        count['世界卫生组织'] += 1
    if '医院' in ru:
        count['医院'] += 1
    if ('公安' in ru) or ('警' in ru):
        count['公安'] += 1
    if '院士' in ru:
        count['院士'] += 1
        
count
```

```python
# 绘制来源占比饼状图
from pyecharts import options as opts
from pyecharts.charts import Page, Pie


def pie_base() -> Pie:
    c = (
        Pie()
        .add("", [list(z) for z in zip(list(count.keys()), list(count.values()))])
        .set_global_opts(title_opts=opts.TitleOpts(title="辟谣来源"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c


c = pie_base()
c.render_notebook()
c.render('辟谣来源占比.html')
```

## 谣言关键词词云

```python
# 通过分词统计词频，结合实际剔除一些没用的词
l = ''.join(list(rumous.title))
import jieba
ls = jieba.lcut(l)
c = {}
for i in ls:
    i = i.replace('?','').replace('？','').replace('病毒','').replace('可以','').replace('新冠','')\
        .replace('武汉','').replace('效果','').replace('家里','').replace('在家','')\
        .replace('的','').replace('，','').replace('人','').replace('有','').replace('和','')\
        .replace('或','')
    c[i] = c.get(i,0) + 1
```

```python
# 绘制词云
from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType


words = list(c.items())
words = sorted(words,key=lambda w: w[1],reverse=True)


def wordcloud_base() -> WordCloud:
    c = (
        WordCloud()
        .add("", words, word_size_range=[20, 100],shape='diamond')
        .set_global_opts(title_opts=opts.TitleOpts(title="谣言关键词词云"))
    )
    return c


c = wordcloud_base()
c.render_notebook()
c.render('谣言词云.html')
```
